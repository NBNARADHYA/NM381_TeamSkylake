```python
files=os.listdir("images")
```

```python
for file in files:
    shutil.move("images/"+file,"/home/rath772k/all_images/")
```

```python
folders = os.listdir("/home/rath772k/all_images")
```

```python
image_names=[]
masks=[]
for folder in folders:
    img_path="/home/rath772k/all_images/"+folder
    mask_path = "/home/rath772k/Masks/"+folder+".png"
    for file in glob.glob(img_path+"/*.jpg"):
        image_names.append(file)
        masks.append(mask_path)
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(list(zip(image_names, masks)), 
               columns =['image', 'mask'])
```
